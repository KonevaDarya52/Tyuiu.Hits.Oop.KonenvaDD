```

public class CreditCard
{
    public string Number { get; set; }
    public string Owner { get; set; }
    protected decimal Money { get; set; }

    public CreditCard(string number, string owner, decimal money = 0)
    {
        Number = number;
        Owner = owner;
        Money = money;
    }

    public virtual string GetInfo()
    {
        return $"Карта: {Number}\nВладелец: {Owner}\nБаланс: {Money} руб.";
    }

    public virtual string Pay(decimal amount)
    {
        if (Money >= amount)
        {
            Money -= amount;
            return $"Оплата {amount} руб. Успешно! Осталось: {Money} руб.";
        }
        else
        {
            return $"Не хватает денег! Нужно: {amount} руб., есть: {Money} руб.";
        }
    }

    public string TransferMoney(CreditCard otherCard, decimal amount)
    {
        if (Money >= amount)
        {
            Money -= amount;
            otherCard.Money += amount;
            return $"Перевод {amount} руб. на карту {otherCard.Number} выполнен!";
        }
        else
        {
            return $"Не хватает денег для перевода!";
        }
    }
}

public class GoldCard : CreditCard
{
    public int Bonuses { get; private set; }

    public GoldCard(string number, string owner, decimal money = 0) 
        : base(number, owner, money)
    {
        Bonuses = 0;
    }

    public override string Pay(decimal amount)
    {
        string result = base.Pay(amount);
        if (result.Contains("Успешно"))
        {
            int newBonuses = (int)(amount / 100);
            Bonuses += newBonuses;
            result += $"\nНачислено бонусов: {newBonuses}";
        }
        return result;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $"\nБонусы: {Bonuses}";
    }
}

public class PremiumCard : CreditCard
{
    public string Support { get; set; }

    public PremiumCard(string number, string owner, string support, decimal money = 0) 
        : base(number, owner, money)
    {
        Support = support;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $"\nПоддержка: {Support}";
    }
}

        Console.WriteLine("=== КРЕДИТНЫЕ КАРТЫ ===\n");

        CreditCard ivanCard = new CreditCard("1111-2222", "Иван", 5000);
        GoldCard petrCard = new GoldCard("3333-4444", "Петр", 10000);
        PremiumCard mariaCard = new PremiumCard("5555-6666", "Мария", "8-800-555", 500);

      
        Console.WriteLine("ИНФОРМАЦИЯ О КАРТАХ:");
        Console.WriteLine(ivanCard.GetInfo());
        Console.WriteLine(petrCard.GetInfo());
        Console.WriteLine(mariaCard.GetInfo());
        Console.WriteLine("----------------------");

        
        Console.WriteLine("\nПЕРЕВОД ДЕНЕГ:");
        Console.WriteLine("Иван переводит Марии 1000 руб:");
        Console.WriteLine(ivanCard.TransferMoney(mariaCard, 1000));
        Console.WriteLine("----------------------");


        Console.WriteLine("\nБАЛАНСЫ ПОСЛЕ ПЕРЕВОДА:");
        Console.WriteLine(ivanCard.GetInfo());
        Console.WriteLine(mariaCard.GetInfo());
        Console.WriteLine("----------------------");

       
        Console.WriteLine("\nПОКУПКИ:");
        Console.WriteLine("Иван покупает на 2000 руб:");
        Console.WriteLine(ivanCard.Pay(2000));
        
        Console.WriteLine("\nПетр покупает на 3000 руб:");
        Console.WriteLine(petrCard.Pay(3000));
        
        Console.WriteLine("\nМария покупает на 800 руб:");
        Console.WriteLine(mariaCard.Pay(800));
        Console.WriteLine("----------------------");

        Console.WriteLine("\nФИНАЛЬНЫЙ БАЛАНС:");
        Console.WriteLine(ivanCard.GetInfo());
        Console.WriteLine(petrCard.GetInfo());
        Console.WriteLine(mariaCard.GetInfo());
```
